```python
x=np.linspace(-10,10,201)
y=4*x+7
a=np.linspace(-10,10,201)
b=np.linspace(-10,10,201)
e=[]
error=[]
i=0
for a1 in a:
    j=0
    for b1 in b:
        e.append([a1,b1,y[j]-a1*x[j]-b1])
        error.append((y[j]-a1*x[j]-b1)*(y[j]-a1*x[j]-b1))
        j=j+1
        i=i+1
for i in range(len(e)):
    print('a=',e[i][0],' b=',e[i][1],' error=',e[i][2])
```

```python
type(error)
```

```python
err=np.array(error).reshape(201,201)
```

```python
sum=0
for e1 in error: 
    sum=sum+e1*e1
avgError = sum/len(error)
print('average error = ',avgError)
```
