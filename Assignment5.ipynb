```python
x = df1['Year']
y = df1['inc']
plt.scatter(x,y)
```

```python
l=0
for i in x:
    l=l+1
sumx=0
sumy=0
sumxy=0
sumxsq=0
for i in x:
    sumx=sumx+i
    sumxsq=sumxsq+i*i
for i in y:
    sumy=sumy+i
for i in range(0,47):
    sumxy=sumxy+x[i]*y[i]
m=(l*sumxy-sumx*sumy)/(l*sumxsq-sumx*sumx)
c=(sumy*sumxsq-sumx*sumxy)/(l*sumxsq-sumx*sumx)
x1=2023
y1=m*x1+c
plt.scatter(x,y,color='red')
plt.plot(x,m*x+c,color='green')
print('for year ',x1, ' the approximation is ',y1)
```
